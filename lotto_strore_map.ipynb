```python

# 크롤링 하는 라인

import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
from tqdm import tqdm

# google 드라이버를 이용해 복권 사이트 정보 가져오기
driver = webdriver.Chrome('c:/chromedriver')
url = 'https://dhlottery.co.kr/store.do?method=topStore&pageGubun=L645'
driver.get(url)
test = []

# data 프레임 생성
df_lotto_stores = pd.DataFrame()

for i in tqdm(range(885, 937)):
    select = Select(driver.find_element_by_id('drwNo')) # 회차를 고르기 위한 버튼
    select.select_by_value(f'{i}') # 회차 선택
    driver.find_element_by_xpath('//*[@id="searchBtn"]').click() # 조회 버튼 클릭
    time.sleep(1)
    new_url = driver.page_source # 새로운 페이지에서 url을 받아오기 위해서 사용
    df = pd.read_html(new_url)
    df_lotto_store = df[1].T
    # 하나의 데이터프레임으로 만들기 위해 합치기
    df_lotto_stores = pd.concat([df_lotto_stores, df_lotto_store], axis=1, ignore_index=True)

driver.close()
pd.set_option('display.max_rows', 10000)
df_all = df_lotto_stores.T[["상호명", "소재지"]]
df_all  
```
